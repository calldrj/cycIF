```python
import pandas as pd
import re

path = './052018_Nuc/Results-Nuc-'
WELLS = ['B0{}'.format(k) for k in range(2, 5)]
FIELD = ['fld{}'.format(k) for k in range(1, 10)]

files_Nuc = [path + '{}{}.txt'.format(w, f) 
              for w in WELLS for f in FIELD]

df_Nuc = [pd.read_csv(file, sep='\t', 
                 engine='python', 
                 usecols=['Label', 'Area', 'IntDen']) 
           for file in files_Nuc]

n = len(df_Nuc)
len(WELLS), len(FIELD), len(df_Nuc)
```

```python
for df in df_Nuc:
    print('\n', '* ' * 26, '\n')
    print(df.head(10))
```

```python
for df in df_Nuc:
    print('\n', '* ' * 26, '\n')
    print(df.info())
```

```python
well = r'[A-Z]+.*?(?=_)'
field = r'(?<=d)[\d]{1,1}?(?=:)'
cell_id = r'(\d+)-(\d+)'
channel = r'(?<=:)[A-Z]+.*?(?=-)'

for k in range(len(df_Nuc)):
    df_Nuc[k]['Well'] = df_Nuc[k]['Label'].apply(lambda x: re.search(well, x).group())
    df_Nuc[k]['Field'] = df_Nuc[k]['Label'].apply(lambda x: re.search(field, x).group())
    df_Nuc[k]['Cell ID'] = df_Nuc[k]['Label'].apply(lambda x: re.search(cell_id, x).group())
    df_Nuc[k]['Channel'] = df_Nuc[k]['Label'].apply(lambda x: re.search(channel, x).group())
    df_Nuc[k]['Cycle'] = df_Nuc[k]['Label'].apply(lambda x: int(x[-1]))
    df_Nuc[k] = df_Nuc[k][['Cell ID', 'Well', 'Field', 'Channel', 'Cycle', 'Area', 'IntDen']]
```

```python
for df in df_Nuc:
    print('\n', '* ' * 32, '\n')
    print(df.head())
```

```python
for df in df_Nuc[:n]:
    print('\n', '* ' * 26, '\n')
    print(df.info())
```
